```python
# Read files
data_size = pd.read_csv('https://raw.githubusercontent.com/amanda-araujo/animal-development/main/Output-Data/size.txt', header = None)
data_length = pd.read_csv('https://raw.githubusercontent.com/amanda-araujo/animal-development/main/Output-Data/length.txt', header = None)
data_width = pd.read_csv('https://raw.githubusercontent.com/amanda-araujo/animal-development/main/Output-Data/width.txt', header = None)

data_size.columns = ["area"]
data_length.columns = ["length"]
data_width.columns = ["width"]
```

# Size measurement

```python
# Data
data = data_size

# Create a list of indices
indices = list(range(len(data)))

# Plot the data
plt.figure(figsize=(18, 10))
plt.rcParams.update({'font.size': 14})
plt.plot(indices, data, c='#ec7569')
plt.scatter(indices, data, s=5)

# Title and labels
plt.title('Larva size')
plt.xlabel('Time step')
plt.ylabel('Area')

# Save figure
plt.savefig("size.png", dpi=300)

# Display the plot
plt.show()
```

## Smoothing

```python
# Data
data = data_size

# Create a list of indices
indices = list(range(len(data)))

# Plot the data
plt.figure(figsize=(18, 10))
plt.rcParams.update({'font.size': 14})
plt.plot(indices, data, c='#ec7569')
plt.scatter(indices, data, s=5)

# Title and labels
plt.title('Larva size')
plt.xlabel('Time step')
plt.ylabel('Area')

import numpy as np
from scipy.signal import savgol_filter

# Smoothing
window_length = 180
polyorder = 1

x = indices #590
y = data    #590

x = np.array(x)
y = np.array(y)

x = np.reshape(x, (len(x), ))
y = np.reshape(y, (len(y), ))

data_filtered = savgol_filter(y, window_length , polyorder) #590

# Plot smoothing
plt.plot(indices, data_filtered, color='blue')

# Save figure
plt.savefig("size_smooth.png", dpi=300)

# Display the plot
plt.show()
```

# Length measurement

# Length & Width

```python
# Length and Width measures together

# Create a list of indices
indices = list(range(len(data)))

# Plot the data
plt.figure(figsize=(18, 10))
plt.rcParams.update({'font.size': 14})
plt.plot(indices, data_length, c='blue')
plt.plot(indices, data_width, c='green')
plt.scatter(indices, data_length, s=5, c='blue')
plt.scatter(indices, data_width, s=5,  c='green')

# Smooth operator
window_length = 180
polyorder = 1

x = indices #590
y = data_length    #590

x = np.array(x)
y = np.array(y)

x = np.reshape(x, (len(x), ))
y = np.reshape(y, (len(y), ))

yhat = savgol_filter(y, window_length , polyorder) #590
plt.plot(x, yhat, c='orange')

y = data_width    #590
y = np.array(y)
y = np.reshape(y, (len(y), ))

yhat = savgol_filter(y, window_length, polyorder) #590
plt.plot(x, yhat, c='red')

# Title and labels
plt.title('Larva growth')
plt.xlabel('Time step')
plt.ylabel('Larva growth')

# Legend for the plot
plt.legend(['Length', 'Width'])

# Save figure
plt.savefig("width_length_smooth.png", dpi=300)

# Display the plot
plt.show()
```
